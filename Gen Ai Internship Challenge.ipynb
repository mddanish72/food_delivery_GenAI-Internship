```python
import sqlite3
import pandas as pd

conn = sqlite3.connect(":memory:")

with open(r"C:\Users\moham\Desktop\assignments\power BI\intership\restaurants.sql", "r") as f:
    sql_script = f.read()

conn.executescript(sql_script)

restaurants = pd.read_sql("SELECT * FROM restaurants", conn)
restaurants.head()
```

```python
# Checking the details

restaurants.info()
```

```python
# Merging the dataset Orders + Users

df = orders.merge(users, on="user_id", how="left")

# Adding Restaurants to it

df = df.merge(restaurants, on="restaurant_id", how="left")
```

```python
# Fixing the date format

df['order_date'] = pd.to_datetime(df['order_date'], dayfirst=True)
```

```python
# Handling duplicate restaurant column names

df['restaurant_name'] = df['restaurant_name_x']

df.drop(columns=['restaurant_name_x', 'restaurant_name_y'],inplace=True)
```

## Solving the MCQs

```python
# City with highest revenue from Gold members

df[df['membership'] == 'Gold'] \
.groupby('city')['total_amount'].sum().idxmax()
```

```python
# Cuisine with highest average order value

df.groupby('cuisine')['total_amount'].mean().idxmax()
```

```python
# How many distinct users placed orders worth more than ₹1000 in total?

#Calculatimg total spend per user
user_total_spend = df.groupby('user_id')['total_amount'].sum()

#Counting users with spend > 1000
distinct_users_gt_1000 = (user_total_spend > 1000).sum()

distinct_users_gt_1000
```

```python
# Which restaurant rating range generated the highest total revenue?

# Creating the rating range
df['rating_range'] = pd.cut(
    df['rating'],
    bins=[3.0, 3.5, 4.0, 4.5, 5.0],
    labels=['3.0–3.5', '3.6–4.0', '4.1–4.5', '4.6–5.0']
)

# Calulating the total revenue per range
rating_revenue = df.groupby('rating_range')['total_amount'].sum()
rating_revenue
```

```python
# Finding the highest among them

rating_revenue.idxmax()
```

```python
# Among Gold members, which city has the highest average order value?

gold_city_aov = (df[df['membership'] == 'Gold'].groupby('city')['total_amount'].mean())
gold_city_aov.idxmax()
```

```python
# Which cuisine has the lowest number of distinct restaurants but still contributes significant revenue?

cuisine_analysis = (
    df.groupby('cuisine')
    .agg(
        distinct_restaurants=('restaurant_id', 'nunique'),
        total_revenue=('total_amount', 'sum')
    )
    .sort_values(
        by=['distinct_restaurants', 'total_revenue'],
        ascending=[True, False]
    )
)
cuisine_analysis.index[0]
```

```python
# What percentage of total orders were placed by Gold members? (Rounded to nearest integer)

# Total number of orders
total_orders = df.shape[0]

# Orders placed by Gold members
gold_orders = df[df['membership'] == 'Gold'].shape[0]

# Percentage of Gold orders
gold_order_percentage = round((gold_orders / total_orders) * 100)

gold_order_percentage
```

```python
# Which restaurant has the highest average order value but less than 20 total orders?

restaurant_analysis = (
    df.groupby('restaurant_name')
    .agg(
        total_orders=('order_id', 'count'),
        avg_order_value=('total_amount', 'mean')
    )
)

# Filtering restaurants with less than 20 orders
filtered_restaurants = restaurant_analysis[
    restaurant_analysis['total_orders'] < 20
]

# Restaurant with highest average order value
filtered_restaurants.sort_values(
    by='avg_order_value',
    ascending=False
).head(1)
```

```python
# Which combination contributes the highest revenue?

combo_revenue = (df.groupby(['membership', 'cuisine'])['total_amount'].sum().sort_values(ascending=False))
combo_revenue.idxmax()
```

```python
# During which quarter of the year is the total revenue highest?

# Checking the date format
df['order_date'] = pd.to_datetime(df['order_date'], dayfirst=True)

# Extracting the quarter from it
df['quarter'] = df['order_date'].dt.quarter

# Calculating the revenue
quarter_revenue = df.groupby('quarter')['total_amount'].sum()

quarter_revenue.idxmax()
```

```python
# How many total orders were placed by users with Gold membership?

# Filtering Gold member orders
gold_orders_df = df[df['membership'] == 'Gold']

# Counting total orders placed by Gold members
total_gold_orders = gold_orders_df.shape[0]

total_gold_orders
```

```python
# What is the total revenue (rounded to nearest integer) generated from orders placed in Hyderabad city?

# Filtering Hyderabad orders
hyderabad_orders = df[df['city'] == 'Hyderabad']

# Calculating total revenue
hyderabad_revenue = hyderabad_orders['total_amount'].sum()

# Rounding to nearest integer
round(hyderabad_revenue)
```

```python
# How many distinct users placed at least one order?

# Count distinct users who placed at least one order
distinct_users = df['user_id'].nunique()

distinct_users
```

```python
# What is the average order value (rounded to 2 decimals) for Gold members?

# Filter Gold member orders
gold_orders = df[df['membership'] == 'Gold']

# Calculate average order value
gold_aov = round(gold_orders['total_amount'].mean(), 2)

gold_aov
```

```python
# How many orders were placed for restaurants with rating ≥ 4.5?

# Filter orders for restaurants with rating >= 4.5
high_rating_orders = df[df['rating'] >= 4.5]

# Count total orders
high_rating_orders_count = high_rating_orders.shape[0]

high_rating_orders_count
```

```python
# How many orders were placed in the top revenue city among Gold members only?

# Filering the Gold members orders
gold_df = df[df['membership'] == 'Gold']

# Filgtering the top revenue gold members
top_gold_city = (gold_df.groupby('city')['total_amount'].sum().idxmax())

# Counting the orders in that city
gold_orders_top_city = gold_df[gold_df['city'] == top_gold_city].shape[0]

gold_orders_top_city
```
